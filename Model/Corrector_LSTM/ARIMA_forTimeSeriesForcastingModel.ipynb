# (1) ARIMA_Model

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from statsmodels.tsa.arima.model import ARIMA

# Load time series data
# data = pd.read_csv('your_timeseries_data.csv')

data = pd.read_csv('/content/cleanData_TCS_25Aug2004_5sep2024_NSE.csv')


# Convert the 'Close' column to numeric values
data['Close'] = pd.to_numeric(data['Close'], errors='coerce')

# Ensure that there are no NaN values after conversion
data.dropna(subset=['Close'], inplace=True)

y = data['Close']


# Fit ARIMA model (p=1, d=1, q=1 as an example)
model = ARIMA(y, order=(1, 1, 1))
model_fit = model.fit()

# Print the summary
print(model_fit.summary())

# Forecast future values
forecast = model_fit.forecast(steps=10)
plt.plot(y)
plt.plot(np.arange(len(y), len(y) + 10), forecast, color='red')
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from statsmodels.tsa.arima.model import ARIMA
from statsmodels.tsa.stattools import adfuller

```

```python
data = pd.read_csv('/content/cleanData_TCS_25Aug2004_5sep2024_NSE.csv')

# Convert 'Date' column to datetime format
data['Date'] = pd.to_datetime(data['Date'], errors='coerce')

# Set 'Date' as the index of the DataFrame
data.set_index('Date', inplace=True)

# Ensure the datetime index has a frequency (e.g., daily)
data = data.asfreq('D')  # 'D' stands for daily frequency

# Convert the 'Close' column to numeric values (if not already numeric)
data['Close'] = pd.to_numeric(data['Close'], errors='coerce')

# Ensure that there are no NaN values in 'Close' after conversion
data.dropna(subset=['Close'], inplace=True)

# Check if the data is stationary using the Augmented Dickey-Fuller test
def test_stationarity(series):
    result = adfuller(series)
    print(f"ADF Statistic: {result[0]}")
    print(f"p-value: {result[1]}")
    if result[1] > 0.05:
        print("The series is non-stationary. Differencing will be applied.")
    else:
        print("The series is stationary.")

# Perform stationarity test on 'Close' column
test_stationarity(data['Close'])

# Apply differencing if the data is non-stationary
# If the p-value from ADF test > 0.05, we difference the data
y = data['Close']
if adfuller(y)[1] > 0.05:
    y = y.diff().dropna()  # Apply differencing

# Fit ARIMA model (p=1, d=1, q=1 as an example)
model = ARIMA(y, order=(1, 1, 1))  # ARIMA(1, 1, 1) is commonly used, adjust as necessary
model_fit = model.fit()

# Print the summary of the fitted model
print(model_fit.summary())

# Forecast the next 5 steps (days)
forecast_steps = 5
forecast_values = model_fit.forecast(steps=forecast_steps)

# Print the forecasted close prices for the next 5 days
print("\nForecasted Close Prices for the Next 5 Days:")
for i, value in enumerate(forecast_values, 1):
    print(f"Day {i}: {value:.2f}")
```

# (5) SARIMA

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from statsmodels.tsa.statespace.sarimax import SARIMAX
from statsmodels.tsa.stattools import adfuller

# Load the dataset (replace with your actual file path)
data = pd.read_csv('/content/cleanData_TCS_25Aug2004_5sep2024_NSE.csv')

# Convert 'Date' column to datetime format
data['Date'] = pd.to_datetime(data['Date'], errors='coerce')

# Set 'Date' as the index of the DataFrame
data.set_index('Date', inplace=True)

# Ensure the datetime index has a frequency (e.g., daily)
data = data.asfreq('D')  # 'D' stands for daily frequency

# Convert the 'Close' column to numeric values (if not already numeric)
data['Close'] = pd.to_numeric(data['Close'], errors='coerce')

# Ensure that there are no NaN values in 'Close' after conversion
data.dropna(subset=['Close'], inplace=True)

# Check if the data is stationary using the Augmented Dickey-Fuller test
def test_stationarity(series):
    result = adfuller(series)
    print(f"ADF Statistic: {result[0]}")
    print(f"p-value: {result[1]}")
    if result[1] > 0.05:
        print("The series is non-stationary. Differencing will be applied.")
    else:
        print("The series is stationary.")

# Perform stationarity test on 'Close' column
test_stationarity(data['Close'])

# Apply differencing if the data is non-stationary
y = data['Close']
if adfuller(y)[1] > 0.05:
    y = y.diff().dropna()  # Apply differencing

# Fit SARIMA model (p=1, d=1, q=1, P=1, D=1, Q=1, S=7 as an example)
# Here, S=7 for weekly seasonality (if daily data with weekly patterns)
sarima_model = SARIMAX(y,
                       order=(1, 1, 1),        # Non-seasonal ARIMA order
                       seasonal_order=(1, 1, 1, 7),  # Seasonal ARIMA order (S=7 for weekly seasonality)
                       enforce_stationarity=False, enforce_invertibility=False)

# Fit the SARIMA model
sarima_model_fit = sarima_model.fit(disp=False)

# Print the summary of the fitted model
print(sarima_model_fit.summary())

# Forecast the next 5 days
forecast_steps = 5
forecast_values = sarima_model_fit.get_forecast(steps=forecast_steps)
forecast_index = pd.date_range(start=data.index[-1] + pd.Timedelta(days=1), periods=forecast_steps, freq='D')

# Print the forecasted close prices for the next 5 days
print("\nForecasted Close Prices for the Next 5 Days:")
for i, value in enumerate(forecast_values.predicted_mean, 1):
    print(f"Day {i}: {value:.2f}")

# Plot the forecasted values along with the historical data
plt.figure(figsize=(10, 6))
plt.plot(data.index, data['Close'], label='Original Data')  # Plot original data
plt.plot(forecast_index, forecast_values.predicted_mean, color='red', label='Forecasted')  # Plot forecasted data

# Add axis labels and a title
plt.xlabel('Time')
plt.ylabel('Close Price')
plt.title('SARIMA Forecast - Close Price')

# Format the x-axis to show years only (e.g., 2004, 2008, 2012)
plt.gca().xaxis.set_major_formatter(plt.matplotlib.dates.DateFormatter('%Y'))  # Show only years
plt.xticks(rotation=45)

# Display the legend
plt.legend()

# Show the plot
plt.show()
```
